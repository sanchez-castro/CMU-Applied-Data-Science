```python
n_visits={}
n_dis={}

for line in open ('ameervisit.txt'):
    VISITID, HOSP, DRG, *desc=line.split()
    n_visits[VISITID]=DRG
    n_dis[DRG]=VISITID
```

```python
n_diseases={}
for dis in n_visits:
    num=n_visits[dis]
    if (num in n_diseases):
        n_diseases[num]+=1
    else:
        n_diseases[num]=1
    
```

```python
def sortdict(d):
    newlist = []
    for key in d:
        val = int(d[key])
        newlist.append([val, key])
    return (sorted(newlist))
            
```

```python
len(sortdict(n_diseases))
```

```python
sortdict(n_diseases)
```

```python
d = sortdict(n_diseases)

fp = open('tmp.csv', 'w')
for (cnt, drg) in d:
    fp.write("%s, %s\n" % (cnt, drg))
fp.close()
```
